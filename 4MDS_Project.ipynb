```python
# Filled the winner column with super over winner and dropped the eliminator column
df_match['winner']=df_match['winner'].fillna(df_match['eliminator'])
df_match.drop(['eliminator'],axis=1, inplace=True)
```

```python
df_match.head()
```

```python
df_match['team1'].value_counts()
```

```python
# team1, team2 and winner column contain multiple names for the same teams
# Resolving it using a dictionary
team_map={
  "Royal Challengers Bangalore":"Royal Challengers Bengaluru",
  "Kings XI Punjab": "Punjab Kings",
  "Delhi Daredevils": "Delhi Capitals",
  "Rising Pune Supergiant": "Rising Pune Supergiants"
}

df_match['team1'] = df_match['team1'].apply(lambda x: team_map[x] if x in team_map.keys() else x)
df_match['team2'] = df_match['team2'].apply(lambda x: team_map[x] if x in team_map.keys() else x)
df_match['winner'] = df_match['winner'].apply(lambda x: team_map[x] if x in team_map.keys() else x)
```

```python
df_match['winner'].value_counts()
```

```python
# Extracting Year from Match Date
df_match['match_date']=pd.to_datetime(df_match['match_date'])
df_match["Year"]=df_match['match_date'].dt.year
```

```python
# Checking for discrepancy in toss decision
df_match['toss_decision'].value_counts()
```

```python
# Checking for discrepancy in result
df_match['result'].value_counts()
```

```python
df_match['winner'].value_counts()
```

```python
df_match.isna().sum()
```

```python
# Examining the null values in city column
df_match[df_match['city'].isna()]['venue'].unique()
```

```python
# Imputing the missing values in city column
city_map ={
    'Dubai International Cricket Stadium': 'Dubai',
    'Sharjah Cricket Stadium': 'Sharjah'
}

df_match['city'] = df_match['city'].fillna(df_match['venue'].apply(lambda x: city_map[x] if x in city_map.keys() else x))
```

```python
# Checking for multiple names of city
df_match['city'].value_counts()
```

```python
# Changing Bangalore to Bengaluru
city_map_2 = {'Bangalore': 'Bengaluru'}
df_match['city'] = df_match['city'].apply(lambda x: city_map_2[x] if x in city_map_2.keys() else x)
```

### Feature Engineering

#### Features related to batsman

```python
# Runs Scored By a Batter
df_ball.groupby("Batter").sum()["BatsmanRun"]
```

```python
# Total Balls Played by a batter
total_balls_played = df_ball.groupby("Batter").count()["BallNumber"]
total_balls_played
```

```python
# Balls played by a batter in a single innings
df_ball[["ID","Batter"]].value_counts()
```

```python
# Innings Played by a batter
batters_df=pd.melt(df_ball,id_vars=["ID"],value_vars=["Batter","NonStriker"],var_name="Role",value_name="Player")
batters_df.groupby("Player")["ID"].nunique().sort_values(ascending=False)
```

```python
innings_grpby = batters_df.groupby("Player")["ID"].nunique()
innings_grpby.index.name = "Batter"
innings_grpby
```

```python
from functools import reduce

dfs=[df_ball.groupby("Batter").sum()["BatsmanRun"],df_ball.groupby("Batter").count()["BallNumber"],batters_df.groupby("Player")["ID"].nunique().sort_values(ascending=False)]

reduced_column=reduce(lambda left,right: pd.merge(left,right,left_index=True,right_index=True,how='outer',),dfs)

reduced_column
```

```python
reduced_column.columns=["Batter_Runs","Balls_Played","Innings_Played"]
reduced_column
```

```python
# No. of runs scored by batsmen (phase wise)
batsman_runs_by_phase = df_ball.groupby(["Batter","Phase"])["BatsmanRun"].sum().reset_index()
batsman_runs_pp = batsman_runs_by_phase[batsman_runs_by_phase["Phase"]=="PowerPlay"].set_index("Batter")[["BatsmanRun"]]
batsman_runs_pp = batsman_runs_pp.rename(columns={"BatsmanRun": "Batter_Runs_PP"})
batsman_runs_middle = batsman_runs_by_phase[batsman_runs_by_phase["Phase"]=="Middle Overs"].set_index("Batter")[["BatsmanRun"]]
batsman_runs_middle = batsman_runs_middle.rename(columns={"BatsmanRun": "Batter_Runs_Middle"})
batsman_runs_death = batsman_runs_by_phase[batsman_runs_by_phase["Phase"]=="Death Overs"].set_index("Batter")[["BatsmanRun"]]
batsman_runs_death = batsman_runs_death.rename(columns={"BatsmanRun": "Batter_Runs_Death"})
```

```python
# Adding these columns
phase_wise_runs = [reduced_column, batsman_runs_pp, batsman_runs_middle, batsman_runs_death]
reduced_column = reduce(lambda left,right: pd.merge(left,right,left_index=True,right_index=True,how='outer',), phase_wise_runs)
reduced_column
```

```python
# Balls Played by Phase
balls_played_by_phase = df_ball.groupby(["Batter","Phase"])["BallNumber"].count().reset_index()
balls_played_pp = balls_played_by_phase[balls_played_by_phase["Phase"]=="PowerPlay"].set_index("Batter")[["BallNumber"]]
balls_played_pp = balls_played_pp.rename(columns={"BallNumber": "Balls_Played_PP"})
balls_played_middle = balls_played_by_phase[balls_played_by_phase["Phase"]=="Middle Overs"].set_index("Batter")[["BallNumber"]]
balls_played_middle = balls_played_middle.rename(columns={"BallNumber": "Balls_Played_Middle"})
balls_played_death = balls_played_by_phase[balls_played_by_phase["Phase"]=="Death Overs"].set_index("Batter")[["BallNumber"]]
balls_played_death = balls_played_death.rename(columns={"BallNumber": "Balls_Played_Death"})
```

```python
# Adding these columns
phase_wise_balls = [reduced_column, balls_played_pp, balls_played_middle, balls_played_death]
reduced_column = reduce(lambda left,right: pd.merge(left,right,left_index=True,right_index=True,how='outer',), phase_wise_balls)
reduced_column
```

```python
# Adding the Batter Strike Rate column.
df_to_be_added=[]
reduced_column["Batter_Strike_Rate"]=round((reduced_column["Batter_Runs"]/reduced_column["Balls_Played"])*100,2)
reduced_column['Batter_Strike_Rate_PP'] = round((reduced_column['Batter_Runs_PP'] / reduced_column['Balls_Played_PP']) * 100, 2)
reduced_column['Batter_Strike_Rate_Middle'] = round((reduced_column['Batter_Runs_Middle'] / reduced_column['Balls_Played_Middle']) * 100, 2)
reduced_column['Batter_Strike_Rate_Death'] = round((reduced_column['Batter_Runs_Death'] / reduced_column['Balls_Played_Death']) * 100, 2)
reduced_column.head()
```

```python
# Calculating the number of times a batter has gotten out
df_ball["PlayerOut"].value_counts()

# Merging it to the current dataset
reduced_column=pd.merge(reduced_column,df_ball["PlayerOut"].value_counts(),left_index=True,right_index=True,how='outer')
reduced_column.head()
```

```python
reduced_column.rename(columns={"count":"Number_of_Dismissals"},inplace=True)
reduced_column.head()
```

```python
# Calculating the Batter's Average
reduced_column["Batter's_Average"] = reduced_column.apply(lambda x: round(x["Batter_Runs"] / x["Number_of_Dismissals"], 2) if pd.notna(x["Number_of_Dismissals"]) or (x["Number_of_Dismissals"] != 0) else np.nan, axis=1)
reduced_column.head()
```

```python
# Dot ball percentage
dot_ball_df = df_ball[df_ball['TotalRun']==0]
dot_balls_played = dot_ball_df.groupby("Batter").count()["BallNumber"]
reduced_column = pd.merge(reduced_column, dot_balls_played, how='outer', left_index=True, right_index=True)
```

```python
reduced_column.rename(columns={"BallNumber": "Dot_Balls_Played"}, inplace=True)
reduced_column.head()
```

```python
reduced_column['Dot_Balls_Percentage'] = round((reduced_column['Dot_Balls_Played'] / reduced_column['Balls_Played']) * 100, 2)
reduced_column.drop('Dot_Balls_Played', axis=1, inplace=True)
reduced_column.head()
```

```python
# Dot Ball Played (Phase Wise)
dot_balls_played_by_phase = dot_ball_df.groupby(["Batter","Phase"]).count()["BallNumber"].reset_index()
dot_balls_played_pp = dot_balls_played_by_phase[dot_balls_played_by_phase["Phase"]=="PowerPlay"].set_index("Batter")[["BallNumber"]]
dot_balls_played_pp = dot_balls_played_pp.rename(columns={"BallNumber": "Dot_Balls_Played_PP"})
dot_balls_played_middle = dot_balls_played_by_phase[dot_balls_played_by_phase["Phase"]=="Middle Overs"].set_index("Batter")[["BallNumber"]]
dot_balls_played_middle = dot_balls_played_middle.rename(columns={"BallNumber": "Dot_Balls_Played_Middle"})
dot_balls_played_death = dot_balls_played_by_phase[dot_balls_played_by_phase["Phase"]=="Death Overs"].set_index("Batter")[["BallNumber"]]
dot_balls_played_death = dot_balls_played_death.rename(columns={"BallNumber": "Dot_Balls_Played_Death"})
```

```python
# Adding these columns
phase_wise_dot_balls = [reduced_column, dot_balls_played_pp, dot_balls_played_middle, dot_balls_played_death]
reduced_column = reduce(lambda left,right: pd.merge(left,right,left_index=True,right_index=True,how='outer',), phase_wise_dot_balls)
reduced_column.head()
```

```python
# Dot Balls Percentage (Phase Wise)
reduced_column['Dot_Balls_Percentage_PP'] = round((reduced_column['Dot_Balls_Played_PP'] * 100 / reduced_column['Balls_Played_PP']), 2)
reduced_column['Dot_Balls_Percentage_Middle'] = round((reduced_column['Dot_Balls_Played_Middle'] * 100 / reduced_column['Balls_Played_Middle']), 2)
reduced_column['Dot_Balls_Percentage_Death'] = round((reduced_column['Dot_Balls_Played_Death'] * 100 / reduced_column['Balls_Played_Death']), 2)
reduced_column.drop(['Dot_Balls_Played_PP', 'Dot_Balls_Played_Middle', 'Dot_Balls_Played_Death'], axis=1, inplace=True)
reduced_column.head()
```

```python
# No of not outs
reduced_column['Not_Outs'] = reduced_column['Innings_Played'] - reduced_column['Number_of_Dismissals']
reduced_column.head()
```

```python
# Boundary %
boundary_runs = df_ball[df_ball['BatsmanRun'].isin([4,6])].groupby('Batter')['BatsmanRun'].sum()
reduced_column = pd.merge(reduced_column, boundary_runs, how='outer', left_index=True, right_index=True)
reduced_column.rename(columns={"BatsmanRun": "Boundary Runs"}, inplace=True)
reduced_column['Boundary_Percentage'] = np.round((reduced_column['Boundary Runs'] / reduced_column['Batter_Runs']) * 100, 2)
reduced_column.drop('Boundary Runs', axis=1, inplace=True)
reduced_column['Boundary_Percentage'].fillna(0, inplace=True)
reduced_column.head()
```

```python
# Boundary Runs (Phase Wise)
boundary_runs_by_phase = df_ball[df_ball['BatsmanRun'].isin([4,6])].groupby(['Batter', 'Phase'])['BatsmanRun'].sum().reset_index()
boundary_runs_pp = boundary_runs_by_phase[boundary_runs_by_phase["Phase"]=="PowerPlay"].set_index("Batter")[["BatsmanRun"]]
boundary_runs_pp = boundary_runs_pp.rename(columns={"BatsmanRun": "Boundary_Runs_PP"})
boundary_runs_middle = boundary_runs_by_phase[boundary_runs_by_phase["Phase"]=="Middle Overs"].set_index("Batter")[["BatsmanRun"]]
boundary_runs_middle = boundary_runs_middle.rename(columns={"BatsmanRun": "Boundary_Runs_Middle"})
boundary_runs_death = boundary_runs_by_phase[boundary_runs_by_phase["Phase"]=="Death Overs"].set_index("Batter")[["BatsmanRun"]]
boundary_runs_death = boundary_runs_death.rename(columns={"BatsmanRun": "Boundary_Runs_Death"})
```

```python
# Adding these columns
phase_wise_boundary_runs = [reduced_column, boundary_runs_pp, boundary_runs_middle, boundary_runs_death]
reduced_column = reduce(lambda left,right: pd.merge(left,right,left_index=True,right_index=True,how='outer',), phase_wise_boundary_runs)
reduced_column.head()
```

```python
# Boundary Percentage (Phase Wise)
reduced_column['Boundary_Percentage_PP'] = round((reduced_column['Boundary_Runs_PP'] / reduced_column['Batter_Runs_PP']) * 100, 2)
reduced_column['Boundary_Percentage_Middle'] = round((reduced_column['Boundary_Runs_Middle'] / reduced_column['Batter_Runs_Middle']) * 100, 2)
reduced_column['Boundary_Percentage_Death'] = round((reduced_column['Boundary_Runs_Death'] / reduced_column['Batter_Runs_Death']) * 100, 2)
reduced_column.drop(['Boundary_Runs_PP', 'Boundary_Runs_Middle', 'Boundary_Runs_Death'], axis=1, inplace=True)
reduced_column.head()
```

```python
fielder_contribution = df_ball[df_ball["FieldersInvolved"].isna()==False].groupby(["FieldersInvolved","Kind"])["ID"].count().reset_index(name='Count')
fielder_contribution
```

```python
# Split multiple fielder names into a list
fielder_contribution["FieldersInvolved"] = fielder_contribution["FieldersInvolved"].str.split(",")

# Explode to get one player per row
exploded_df = fielder_contribution.explode("FieldersInvolved")

# Remove any whitespaces in names
exploded_df["FieldersInvolved"] = exploded_df["FieldersInvolved"].str.strip()

# Group again to get total contribution per player and kind
final_contrib = exploded_df.groupby(["FieldersInvolved", "Kind"])["Count"].sum().unstack(fill_value=0)

# Display final DataFrame
final_contrib.head()
```

```python
# Calculating the number of outliers per column
def outlier_calculation(series):
  q1=np.percentile(series,25)
  q3=np.percentile(series,75)
  iqr=q3-q1
  lower_bound=q1-(1.5*iqr)
  upper_bound=q3+(1.5*iqr)

  outliers=series[(series<lower_bound) | (series>upper_bound)]
  return len(outliers)

outlier_counts={x:outlier_calculation(reduced_column[x]) for x in reduced_column.columns if reduced_column[x].dtype=="float64"}
outlier_counts=pd.Series(outlier_counts,name="Outlier_Count")
outlier_counts
```

```python
# Getting the names of the columns with ad without outliers
columns_with_outliers=[x for x,y in outlier_counts.items() if y>0]
columns_without_outliers=[x for x,y in outlier_counts.items() if y==0]
print("Columns with outliers are: ",columns_with_outliers)
print("Columns without outliers are: ",columns_without_outliers)
```

```python
def sort_dataframes():
    """Sort all dataframes according to their specific performance criteria"""

    global good_PP_batters_kmeans, good_middle_batters_kmeans, good_death_overs_hitters_kmeans
    global good_T20_opening_bowlers_kmeans, Middle_T20_Bowlers_cluster_kmeans, death_T20_bowlers_kmeans

    try:
        # PowerPlay Batters: Sort by Batter_Runs_PP (desc) then Batter_Strike_Rate_PP (desc)
        good_PP_batters_kmeans = good_PP_batters_kmeans.sort_values(
            by=['Batter_Runs_PP', 'Batter_Strike_Rate_PP'],
            ascending=[False, False]
        )

        # Middle Overs Batters: Sort by Batter_Runs_Middle (desc) then Batter's_Average (desc)
        good_middle_batters_kmeans = good_middle_batters_kmeans.sort_values(
            by=['Batter_Runs_Middle', "Batter's_Average"],
            ascending=[False, False]
        )

        # Death Overs Hitters: Sort by Batter_Runs_Death (desc) then Batter_Strike_Rate_Death (desc)
        good_death_overs_hitters_kmeans = good_death_overs_hitters_kmeans.sort_values(
            by=['Batter_Runs_Death', 'Batter_Strike_Rate_Death'],
            ascending=[False, False]
        )

        # Opening Bowlers: Sort by Wickets_Taken_PP (desc) then Bowling_Economy_Rate_PP (asc)
        good_T20_opening_bowlers_kmeans = good_T20_opening_bowlers_kmeans.sort_values(
            by=['Wickets_Taken_PP', 'Bowling_Economy_Rate_PP'],
            ascending=[False, True]
        )

        # Middle Overs Bowlers: Sort by Wickets_Taken_Middle (desc), Economy (asc), then Average (asc)
        Middle_T20_Bowlers_cluster_kmeans = Middle_T20_Bowlers_cluster_kmeans.sort_values(
            by=['Wickets_Taken_Middle', 'Bowling_Economy_Rate_Middle', 'Bowling_Average'],
            ascending=[False, True, True]
        )

        # Death Overs Bowlers: Sort by Wickets_Taken_Death (desc) then Bowling_Economy_Rate_Death (asc)
        death_T20_bowlers_kmeans = death_T20_bowlers_kmeans.sort_values(
            by=['Wickets_Taken_Death', 'Bowling_Economy_Rate_Death'],
            ascending=[False, True]
        )

        print(" All dataframes sorted successfully!")

    except Exception as e:
        print(f" Error sorting dataframes: {e}")
        print("Using original dataframe order...")

sort_dataframes()

def welcome_html():
    import base64

    def encode_image(path):
        with open(path, "rb") as f:
            return base64.b64encode(f.read()).decode()

    # Logo
    logo_b64 = encode_image("/content/drive/My Drive/PROJECT (4MDS)/logo.jpg")

    # Team member images (replace with your actual files)
    members = [
        ("Aayush Joshi", encode_image("/content/drive/My Drive/PROJECT (4MDS)/aayush.jpg")),
        ("Aryan Singh", encode_image("/content/drive/My Drive/PROJECT (4MDS)/aryan.jpg")),
        ("Kartikeya Sinha", encode_image("/content/drive/My Drive/PROJECT (4MDS)/kartikeya.jpg")),
        ("Simran Kaur", encode_image("/content/drive/My Drive/PROJECT (4MDS)/simran.jpg")),
    ]

    # Build team section HTML
    team_html = "<div style='display:flex; justify-content:center; gap:50px; flex-wrap:wrap; margin-top:40px;'>"
    for name, img_b64 in members:
        team_html += f"""
        <div style='text-align:center;'>
          <img src="data:image/jpeg;base64,{img_b64}"
               style="width:110px; height:110px; border-radius:50%; object-fit:cover; box-shadow:0 5px 14px rgba(0,0,0,0.18);" />
          <p style='margin-top:8px; color:#1E3A8A; font-weight:600; font-size:15px;'>{name}</p>
        </div>
        """
    team_html += "</div>"


    return f"""
    <div style='padding:6px 36px 36px; text-align:center'>
      <img src="data:image/jpeg;base64,{logo_b64}"
           style="width:120px; height:120px; border-radius:16px; margin:0 auto 20px; display:block; box-shadow:0 6px 16px rgba(0,0,0,0.15);" />
      <h1 style='color:#1E3A8A; margin:14px 0 6px'>Cricklytics</h1>
      <h3 style='color:#4B5563; margin:0 0 12px'>Player Segmentation & Team Recommendations</h3>
      <p style='color:#6B7280; max-width:720px; margin:0 auto'>
        Data-driven insights for smarter selection. Explore clusters, compare roles, and build a balanced Playing XI.
      </p>
      <div style='margin-top:28px; text-align:left; max-width:920px; margin-inline:auto'>
        <h3 style='color:#1E3A8A'>About Us</h3>
        <p style='color:#374151'>Cricklytics blends ML with cricket domain knowledge to deliver explainable player segmentation and transparent team recommendations.</p>
        <h3 style='color:#1E3A8A; margin-top:18px'>Services Offered</h3>
        <ul style='color:#374151; line-height:1.7'>
          <li>Player Insights - Explore player statistics and compare players side-by-side across performance metrics.</li>
          <li>Team Recommendation - Generate data-driven, role-balanced XI suggestions with customizable filters.</li>
        </ul>
      </div>

      <div style='margin-top:50px;'>
        <h3 style='color:#1E3A8A; margin-bottom:24px;'>Meet the Team</h3>
        {team_html}
      </div>
    </div>
    """

def create_home_page():
    """Create the home page interface using the welcome_html function"""
    with gr.Column() as home_interface:
        gr.HTML(welcome_html())
    return home_interface

def create_player_overview():
    """Create the Player Overview tab with a single searchable dropdown for player selection."""

    # Load your actual data
    try:
        player_stats_df = reduced_column.copy()
        player_names = player_stats_df.index.tolist()
    except:
        # Fallback with sample data based on your dataset structure
        player_names = ['V Kohli', 'RG Sharma', 'MS Dhoni', 'AB de Villiers', 'SK Raina', 'KA Pollard', 'RA Jadeja']
        player_stats_df = None

    def get_player_stats(player_name):
        """Get comprehensive stats for selected player from your actual data"""
        if not player_name:
            return "Please select a player to view their statistics."

        try:
            if player_stats_df is not None:
                player_data = player_stats_df.loc[player_name]

                # Build comprehensive stats HTML (same as before)
                stats_html = f"""
                <div style="background: #f8fafc; padding: 2rem; border-radius: 10px; margin: 1rem 0;">
                    <h2 style="color: #1E3A8A; text-align: center; margin-bottom: 2rem;"> {player_name} - Career Statistics</h2>
                    <div style="display: grid; grid-template-columns: repeat(5, 1fr); gap: 1rem; margin-bottom: 2rem;">
                        <!-- Batting -->
                        <div style="background: white; padding: 1.5rem; border-radius: 8px; border-left: 4px solid #1E3A8A;">
                            <h3 style="color: #1E3A8A; margin-bottom: 1rem; font-size: 0.9rem;"> Overall Batting Performance</h3>
                            <p style="font-size: 0.85rem;"><strong>Total Runs:</strong> {player_data.get('Batter_Runs', 0):,.0f}</p>
                            <p style="font-size: 0.85rem;"><strong>Strike Rate:</strong> {player_data.get('Batter_Strike_Rate', 0):.2f}</p>
                            <p style="font-size: 0.85rem;"><strong>Average:</strong> {player_data.get("Batter's_Average", 0):.2f}</p>
                            <p style="font-size: 0.85rem;"><strong>Innings Played:</strong> {player_data.get('Innings_Played', 0):.0f}</p>
                            <p style="font-size: 0.85rem;"><strong>30+ Scores:</strong> {player_data.get('30+ scores', 0):.0f}</p>
                            <p style="font-size: 0.85rem;"><strong>Highest Score:</strong> {player_data.get('Highest Score', 0):.0f}</p>
                        </div>
                        <!-- Phase-wise Batting -->
                        <div style="background: white; padding: 1.5rem; border-radius: 8px; border-left: 4px solid #059669;">
                            <h3 style="color: #059669; margin-bottom: 1rem; font-size: 0.9rem;"> Phase-wise Batting</h3>
                            <p style="font-size: 0.85rem;"><strong>PowerPlay SR:</strong> {player_data.get('Batter_Strike_Rate_PP', 0):.2f}</p>
                            <p style="font-size: 0.85rem;"><strong>Middle Overs SR:</strong> {player_data.get('Batter_Strike_Rate_Middle', 0):.2f}</p>
                            <p style="font-size: 0.85rem;"><strong>Death Overs SR:</strong> {player_data.get('Batter_Strike_Rate_Death', 0):.2f}</p>
                            <p style="font-size: 0.85rem;"><strong>Boundary %:</strong> {player_data.get('Boundary_Percentage', 0):.1f}%</p>
                            <p style="font-size: 0.85rem;"><strong>Dot Ball %:</strong> {player_data.get('Dot_Balls_Percentage', 0):.1f}%</p>
                        </div>
                        <!-- Bowling -->
                        <div style="background: white; padding: 1.5rem; border-radius: 8px; border-left: 4px solid #DC2626;">
                            <h3 style="color: #DC2626; margin-bottom: 1rem; font-size: 0.9rem;"> Overall Bowling Performance</h3>
                            <p style="font-size: 0.85rem;"><strong>Wickets:</strong> {player_data.get('Wickets_Taken', 0):.0f}</p>
                            <p style="font-size: 0.85rem;"><strong>Economy Rate:</strong> {player_data.get('Bowling_Economy_Rate', 0):.2f}</p>
                            <p style="font-size: 0.85rem;"><strong>Bowling Average:</strong> {player_data.get('Bowling_Average', 0):.2f}</p>
                            <p style="font-size: 0.85rem;"><strong>Strike Rate:</strong> {player_data.get('Bowling_Strike_Rate', 0):.2f}</p>
                            <p style="font-size: 0.85rem;"><strong>Best Figures:</strong> {player_data.get("Best_Wickets'_Figure", 0):.0f} wickets</p>
                            <p style="font-size: 0.85rem;"><strong>3+ Wicket Hauls:</strong> {player_data.get('3+_Wickets_Haul', 0):.0f}</p>
                        </div>
                        <!-- Phase-wise Bowling -->
                        <div style="background: white; padding: 1.5rem; border-radius: 8px; border-left: 4px solid #1E3A8A;">
                            <h3 style="color: #1E3A8A; margin-bottom: 1rem; font-size: 0.9rem;"> Phase-wise Bowling</h3>
                            <p style="font-size: 0.85rem;"><strong>PowerPlay Economy:</strong> {player_data.get('Bowling_Economy_Rate_PP', 0):.2f}</p>
                            <p style="font-size: 0.85rem;"><strong>Middle Overs Economy:</strong> {player_data.get('Bowling_Economy_Rate_Middle', 0):.2f}</p>
                            <p style="font-size: 0.85rem;"><strong>Death Overs Economy:</strong> {player_data.get("Bowling_Economy_Rate_Death", 0):.2f}</p>
                        </div>
                        <!-- Fielding -->
                        <div style="background: white; padding: 1.5rem; border-radius: 8px; border-left: 4px solid #7C3AED;">
                            <h3 style="color: #7C3AED; margin-bottom: 1rem; font-size: 0.9rem;"> Fielding & Career</h3>
                            <p style="font-size: 0.85rem;"><strong>Catches:</strong> {player_data.get('Catches Taken', 0):.0f}</p>
                            <p style="font-size: 0.85rem;"><strong>Run Outs:</strong> {player_data.get('Run Outs', 0):.0f}</p>
                            <p style="font-size: 0.85rem;"><strong>Stumpings:</strong> {player_data.get('Stumpings', 0):.0f}</p>
                            <p style="font-size: 0.85rem;"><strong>Seasons Played:</strong> {player_data.get('No_of_Seasons_Played', 0):.0f}</p>
                        </div>
                    </div>
                </div>
                """
                return stats_html
            else:
                return f"""
                <div style="background: #f8fafc; padding: 2rem; border-radius: 10px; margin: 1rem 0; text-align: center;">
                    <h2 style="color: #1E3A8A;"> {player_name}</h2>
                    <p>Player statistics will be loaded from your dataset...</p>
                    <p style="color: #6B7280;">Please ensure your player_statistics.csv file is in the correct path.</p>
                </div>
                """
        except Exception as e:
            return f"""
            <div style="background: #fef2f2; padding: 2rem; border-radius: 10px; margin: 1rem 0; text-align: center; border-left: 4px solid #DC2626;">
                <h3 style="color: #DC2626;"> Error Loading Player Data</h3>
                <p>Could not find data for <strong>{player_name}</strong></p>
                <p style="color: #6B7280; font-size: 0.9rem;">Error: {str(e)}</p>
            </div>
            """

    with gr.Column() as player_overview:

        # Header
        gr.HTML("""
        <div style="text-align: center; padding: 2rem; background: #f0f9ff; color: white; border-radius: 10px; margin-bottom: 2rem;">
            <h1> Player Overview</h1>
            <p>Search and explore comprehensive player statistics from your IPL dataset</p>
        </div>
        """)

        # Single Searchable Dropdown Interface
        player_dropdown = gr.Dropdown(
            choices=player_names,
            label=" Search & Select Player",
            info="Type or select a player to view detailed statistics",
            filterable=True
        )

        # Quick Stats Info
        gr.HTML(f"""
        <div style="background: #f0f9ff; padding: 1rem; border-radius: 8px; margin: 1rem 0; text-align: center;">
            <p style="color: #0369a1; margin: 0;">
                 <strong>{len(player_names)} players</strong> available in your dataset |
                 Complete batting, bowling, and fielding statistics |
                 ML-based player clustering included
            </p>
        </div>
        """)

        # Player Statistics Display
        stats_display = gr.HTML(
            value="""
            <div style='text-align: center; padding: 3rem; color: #6B7280; background: #f9fafb; border-radius: 10px; border: 2px dashed #d1d5db;'>
                <div style='font-size: 3rem; margin-bottom: 1rem;'></div>
                <h3 style='color: #374151; margin-bottom: 1rem;'>Select a Player Above</h3>
                <p style='margin: 0;'>Choose any player from the dropdown to view their comprehensive IPL statistics,<br>including batting, bowling, fielding performance, and advanced analytics.</p>
            </div>
            """
        )

        # Dropdown change triggers stats display (single event handler now)
        player_dropdown.change(
            fn=get_player_stats,
            inputs=[player_dropdown],
            outputs=[stats_display]
        )

    return player_overview

def create_player_comparison():
    """Create the Player Comparison tab with player selection and comprehensive comparison including radar charts"""

    try:
        # Load player data
        player_stats_df = reduced_column.copy()
        player_names = player_stats_df.index.tolist()
    except:
        # Fallback with sample data
        player_names = ['V Kohli', 'RG Sharma', 'MS Dhoni', 'AB de Villiers', 'SK Raina']
        player_stats_df = None

    def create_batting_radar_chart(player1, player2):
        """Create a radar chart comparing two players on batting metrics"""
        if not player1 or not player2 or player1 == player2 or player_stats_df is None:
            return None

        try:
            import plotly.graph_objects as go
            import numpy as np
            import pandas as pd

            player1_data = player_stats_df.loc[player1]
            player2_data = player_stats_df.loc[player2]

            # Define batting metrics for radar chart
            metrics = [
                'Batter_Runs', "Batter's_Average", 'Batter_Strike_Rate',
                'Boundary_Percentage', 'Dot_Balls_Percentage', '30+ scores'
            ]

            metric_labels = [
                'Runs', 'Batting Average', 'Batting Strike Rate',
                'Boundary %', 'Dot Ball %', '30+ Scores'
            ]

            # Get dataset-wide min/max for normalization
            dataset_stats = {}
            for metric in metrics:
                values = player_stats_df[metric].dropna()
                dataset_stats[metric] = {
                    'min': values.min(),
                    'max': values.max(),
                    'range': values.max() - values.min()
                }

            def normalize_metric(metric, value):
                if pd.isna(value) or value == 0:
                    return 0

                stats = dataset_stats[metric]
                if stats['range'] == 0:
                    return 50  # middle value if no variance

                # Normalize to 0-100 scale
                if metric in ['Dot_Balls_Percentage']:
                    # Lower is better - invert the scale
                    normalized = (stats['max'] - value) / stats['range'] * 100
                else:
                    # Higher is better
                    normalized = (value - stats['min']) / stats['range'] * 100

                return max(0, min(100, normalized))

            # Get normalized values
            p1_values = [normalize_metric(metric, player1_data.get(metric, 0)) for metric in metrics]
            p2_values = [normalize_metric(metric, player2_data.get(metric, 0)) for metric in metrics]

            # Create radar chart
            fig = go.Figure()

            fig.add_trace(go.Scatterpolar(
                r=p1_values,
                theta=metric_labels,
                fill='toself',
                name=player1,
                line=dict(color='#1E3A8A', width=2),
                fillcolor='rgba(30, 58, 138, 0.25)'
            ))

            fig.add_trace(go.Scatterpolar(
                r=p2_values,
                theta=metric_labels,
                fill='toself',
                name=player2,
                line=dict(color='#DC2626', width=2),
                fillcolor='rgba(220, 38, 38, 0.25)'
            ))

            fig.update_layout(
                polar=dict(
                    radialaxis=dict(
                        visible=True,
                        range=[0, 100],
                        showticklabels=False,
                        ticks="",
                        tickfont=dict(size=10)
                    ),
                    angularaxis=dict(
                        tickfont=dict(size=11)
                    )
                ),
                showlegend=True,
                title=dict(
                    text=f"Batting Comparison: {player1} vs {player2}",
                    x=0.5,
                    font=dict(size=16)
                ),
                height=500,
                width=600
            )

            return fig

        except Exception as e:
            print(f"Error creating batting radar chart: {e}")
            return None

    def create_bowling_radar_chart(player1, player2):
        """Create a radar chart comparing two players on bowling metrics"""
        if not player1 or not player2 or player1 == player2 or player_stats_df is None:
            return None

        try:
            import plotly.graph_objects as go
            import numpy as np
            import pandas as pd

            player1_data = player_stats_df.loc[player1]
            player2_data = player_stats_df.loc[player2]

            # Define bowling metrics for radar chart
            metrics = [
                'Wickets_Taken', 'Bowling_Economy_Rate', 'Bowling_Strike_Rate',
                'Bowling_Average', '3+_Wickets_Haul'
            ]

            metric_labels = [
                'Wickets', 'Economy Rate', 'Bowling Strike Rate',
                'Bowling Average', '3+ Wicket Hauls'
            ]

            # Get dataset-wide min/max for normalization
            dataset_stats = {}
            for metric in metrics:
                values = player_stats_df[metric].dropna()
                dataset_stats[metric] = {
                    'min': values.min(),
                    'max': values.max(),
                    'range': values.max() - values.min()
                }

            def normalize_metric(metric, value):
                if pd.isna(value) or value == 0:
                    return 0

                stats = dataset_stats[metric]
                if stats['range'] == 0:
                    return 50  # middle value if no variance

                # Normalize to 0-100 scale
                if metric in ['Bowling_Economy_Rate', 'Bowling_Average', 'Bowling_Strike_Rate']:
                    # Lower is better - invert the scale
                    normalized = (stats['max'] - value) / stats['range'] * 100
                else:
                    # Higher is better
                    normalized = (value - stats['min']) / stats['range'] * 100

                return max(0, min(100, normalized))

            # Get normalized values
            p1_values = [normalize_metric(metric, player1_data.get(metric, 0)) for metric in metrics]
            p2_values = [normalize_metric(metric, player2_data.get(metric, 0)) for metric in metrics]

            # Create radar chart
            fig = go.Figure()

            fig.add_trace(go.Scatterpolar(
                r=p1_values,
                theta=metric_labels,
                fill='toself',
                name=player1,
                line=dict(color='#1E3A8A', width=2),
                fillcolor='rgba(30, 58, 138, 0.25)'
            ))

            fig.add_trace(go.Scatterpolar(
                r=p2_values,
                theta=metric_labels,
                fill='toself',
                name=player2,
                line=dict(color='#DC2626', width=2),
                fillcolor='rgba(220, 38, 38, 0.25)'
            ))

            fig.update_layout(
                polar=dict(
                    radialaxis=dict(
                        visible=True,
                        range=[0, 100],
                        showticklabels=False,
                        ticks="",
                        tickfont=dict(size=10)
                    ),
                    angularaxis=dict(
                        tickfont=dict(size=11)
                    )
                ),
                showlegend=True,
                title=dict(
                    text=f"Bowling Comparison: {player1} vs {player2}",
                    x=0.5,
                    font=dict(size=16)
                ),
                height=500,
                width=600
            )

            return fig

        except Exception as e:
            print(f"Error creating bowling radar chart: {e}")
            return None

    def compare_players(player1, player2):
        """Compare two selected players and return comprehensive comparison"""
        if not player1 or not player2:
            return """
            <div style='text-align: center; padding: 3rem; color: #6B7280; background: #f9fafb; border-radius: 10px; border: 2px dashed #d1d5db;'>
                <div style='font-size: 3rem; margin-bottom: 1rem;'></div>
                <h3 style='color: #374151; margin-bottom: 1rem;'>Select Two Players to Compare</h3>
                <p style='margin: 0;'>Choose two players from the dropdowns above to see their detailed comparison.</p>
            </div>
            """, None, None

        if player1 == player2:
            return """
            <div style='text-align: center; padding: 2rem; color: #DC2626; background: #fef2f2; border-radius: 10px; border: 1px solid #fecaca;'>
                <h3 style='color: #DC2626;'> Same Player Selected</h3>
                <p>Please select two different players for comparison.</p>
            </div>
            """, None, None

        try:
            if player_stats_df is not None:
                player1_data = player_stats_df.loc[player1]
                player2_data = player_stats_df.loc[player2]

                # Create side-by-side comparison (same HTML as before)
                comparison_html = f"""
                <div style="background: #f8fafc; padding: 2rem; border-radius: 10px; margin: 1rem 0;">
                    <h3 style="color: #1E3A8A; text-align: center; margin-bottom: 2rem;"> {player1} vs {player2}</h3>

                    <!-- Overall Performance Comparison -->
                    <div style="display: grid; grid-template-columns: 1fr 1fr; gap: 2rem; margin-bottom: 2rem;">

                        <!-- Player 1 Stats -->
                        <div style="background: white; padding: 1.5rem; border-radius: 8px; border-left: 4px solid #1E3A8A;">
                            <h4 style="color: #1E3A8A; text-align: center; margin-bottom: 1rem;"> {player1}</h4>

                            <div style="margin-bottom: 1.5rem;">
                                <h5 style="color: #059669; margin-bottom: 0.5rem;">Batting Performance</h5>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Total Runs:</strong> {player1_data.get('Batter_Runs', 0):,.0f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Innings:</strong> {player1_data.get('Innings_Played', 0):.0f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Strike Rate:</strong> {player1_data.get('Batter_Strike_Rate', 0):.2f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Average:</strong> {player1_data.get("Batter's_Average", 0):.2f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>30+ Scores:</strong> {player1_data.get('30+ scores', 0):.0f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Highest Score:</strong> {player1_data.get('Highest Score', 0):.0f}</p>
                            </div>

                            <div style="margin-bottom: 1.5rem;">
                                <h5 style="color: #DC2626; margin-bottom: 0.5rem;">Bowling Performance</h5>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Wickets:</strong> {player1_data.get('Wickets_Taken', 0):.0f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Economy:</strong> {player1_data.get('Bowling_Economy_Rate', 0):.2f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Bowling Avg:</strong> {player1_data.get('Bowling_Average', 0):.2f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Bowling Strike Rate:</strong> {player1_data.get('Bowling_Strike_Rate', 0):.2f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Max Wickets (Match):</strong> {player1_data.get("Max_Wickets_Taken", 0):.0f}</p>
                            </div>

                            <div>
                                <h5 style="color: #7C3AED; margin-bottom: 0.5rem;">Career & Fielding</h5>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Seasons:</strong> {player1_data.get('No_of_Seasons_Played', 0):.0f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Catches:</strong> {player1_data.get('Catches Taken', 0):.0f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Run Outs:</strong> {player1_data.get('Run Outs', 0):.0f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Stumpings:</strong> {player1_data.get('Stumpings', 0):.0f}</p>
                            </div>
                        </div>

                        <!-- Player 2 Stats -->
                        <div style="background: white; padding: 1.5rem; border-radius: 8px; border-left: 4px solid #DC2626;">
                            <h4 style="color: #DC2626; text-align: center; margin-bottom: 1rem;"> {player2}</h4>

                            <div style="margin-bottom: 1.5rem;">
                                <h5 style="color: #059669; margin-bottom: 0.5rem;">Batting Performance</h5>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Total Runs:</strong> {player2_data.get('Batter_Runs', 0):,.0f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Innings:</strong> {player2_data.get('Innings_Played', 0):.0f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Strike Rate:</strong> {player2_data.get('Batter_Strike_Rate', 0):.2f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Average:</strong> {player2_data.get("Batter's_Average", 0):.2f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>30+ Scores:</strong> {player2_data.get('30+ scores', 0):.0f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Highest Score:</strong> {player2_data.get('Highest Score', 0):.0f}</p>
                            </div>

                            <div style="margin-bottom: 1.5rem;">
                                <h5 style="color: #DC2626; margin-bottom: 0.5rem;">Bowling Performance</h5>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Wickets:</strong> {player2_data.get('Wickets_Taken', 0):.0f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Economy:</strong> {player2_data.get('Bowling_Economy_Rate', 0):.2f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Bowling Avg:</strong> {player2_data.get('Bowling_Average', 0):.2f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Bowling Strike Rate:</strong> {player2_data.get('Bowling_Strike_Rate', 0):.2f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Max Wickets (Match):</strong> {player2_data.get("Max_Wickets_Taken", 0):.0f}</p>
                            </div>

                            <div>
                                <h5 style="color: #7C3AED; margin-bottom: 0.5rem;">Career & Fielding</h5>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Seasons:</strong> {player2_data.get('No_of_Seasons_Played', 0):.0f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Catches:</strong> {player2_data.get('Catches Taken', 0):.0f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Run Outs:</strong> {player1_data.get('Run Outs', 0):.0f}</p>
                                <p style="font-size: 0.9rem; margin: 0.25rem 0;"><strong>Stumpings:</strong> {player2_data.get('Stumpings', 0):.0f}</p>
                            </div>
                        </div>
                    </div>

                    <!-- Phase-wise Performance Comparison -->
                    <div style="background: white; padding: 1.5rem; border-radius: 8px; border-left: 4px solid #059669; margin-bottom: 2rem;">
                        <h4 style="color: #059669; margin-bottom: 1rem;"> Phase-wise Performance Comparison</h4>

                        <div style="display: grid; grid-template-columns: 1fr 1fr; gap: 2rem;">
                            <div>
                                <h5 style="color: #1E3A8A; margin-bottom: 0.5rem;">{player1} - Phase Stats</h5>
                                <p style="font-size: 0.85rem; margin: 0.2rem 0;"><strong>PowerPlay SR:</strong> {player1_data.get('Batter_Strike_Rate_PP', 0):.1f}</p>
                                <p style="font-size: 0.85rem; margin: 0.2rem 0;"><strong>Middle Overs SR:</strong> {player1_data.get('Batter_Strike_Rate_Middle', 0):.1f}</p>
                                <p style="font-size: 0.85rem; margin: 0.2rem 0;"><strong>Death Overs SR:</strong> {player1_data.get('Batter_Strike_Rate_Death', 0):.1f}</p>
                                <p style="font-size: 0.85rem; margin: 0.2rem 0;"><strong>PowerPlay Eco:</strong> {player1_data.get('Bowling_Economy_Rate_PP', 0):.1f}</p>
                                <p style="font-size: 0.85rem; margin: 0.2rem 0;"><strong>Middle Overs Eco:</strong> {player1_data.get('Bowling_Economy_Rate_Middle', 0):.1f}</p>
                                <p style="font-size: 0.85rem; margin: 0.2rem 0;"><strong>Death Overs Eco:</strong> {player1_data.get('Bowling_Economy_Rate_Death', 0):.1f}</p>
                            </div>
                            <div>
                                <h5 style="color: #DC2626; margin-bottom: 0.5rem;">{player2} - Phase Stats</h5>
                                <p style="font-size: 0.85rem; margin: 0.2rem 0;"><strong>PowerPlay SR:</strong> {player2_data.get('Batter_Strike_Rate_PP', 0):.1f}</p>
                                <p style="font-size: 0.85rem; margin: 0.2rem 0;"><strong>Middle Overs SR:</strong> {player2_data.get('Batter_Strike_Rate_Middle', 0):.1f}</p>
                                <p style="font-size: 0.85rem; margin: 0.2rem 0;"><strong>Death Overs SR:</strong> {player2_data.get('Batter_Strike_Rate_Death', 0):.1f}</p>
                                <p style="font-size: 0.85rem; margin: 0.2rem 0;"><strong>PowerPlay Eco:</strong> {player2_data.get('Bowling_Economy_Rate_PP', 0):.1f}</p>
                                <p style="font-size: 0.85rem; margin: 0.2rem 0;"><strong>Middle Overs Eco:</strong> {player2_data.get('Bowling_Economy_Rate_Middle', 0):.1f}</p>
                                <p style="font-size: 0.85rem; margin: 0.2rem 0;"><strong>Death Overs Eco:</strong> {player2_data.get('Bowling_Economy_Rate_Death', 0):.1f}</p>
                            </div>
                        </div>
                    </div>
                </div>
                """

                # Create both radar charts
                batting_radar_fig = create_batting_radar_chart(player1, player2)
                bowling_radar_fig = create_bowling_radar_chart(player1, player2)

                return comparison_html, batting_radar_fig, bowling_radar_fig

            else:
                return """
                <div style="text-align: center; padding: 2rem; color: #DC2626;">
                    <h3> Player data not available</h3>
                    <p>Please ensure your player statistics dataset is loaded correctly.</p>
                </div>
                """, None, None

        except Exception as e:
            return f"""
            <div style="text-align: center; padding: 2rem; color: #DC2626;">
                <h3> Error comparing players</h3>
                <p>Error: {str(e)}</p>
            </div>
            """, None, None

    with gr.Column() as player_comparison:

        # Header Section
        gr.HTML("""
        <div style="text-align: center; padding: 2rem; background: #f0f9ff; color: white; border-radius: 10px; margin-bottom: 2rem;">
            <h1> Player Comparison</h1>
            <p>Compare any two players side-by-side across all performance metrics</p>
        </div>
        """)

        # Player Selection Interface
        with gr.Row():
            player1_dropdown = gr.Dropdown(
                choices=player_names,
                label=" Select Player 1",
                info="Choose the first player for comparison",
                filterable=True
            )

            player2_dropdown = gr.Dropdown(
                choices=player_names,
                label=" Select Player 2",
                info="Choose the second player for comparison",
                filterable=True
            )

        # Quick tip
        gr.HTML("""
        <div style="background: #f0f9ff; padding: 1rem; border-radius: 8px; margin: 1rem 0; text-align: center;">
            <p style="color: #0369a1; margin: 0; font-size: 0.9rem;">
                 <strong>Tip:</strong> Select two different players above to see their comprehensive comparison including batting, bowling, and fielding statistics.
            </p>
        </div>
        """)

        # Side-by-Side Comparison Display
        comparison_display = gr.HTML(
            value="""
            <div style='text-align: center; padding: 3rem; color: #6B7280; background: #f9fafb; border-radius: 10px; border: 2px dashed #d1d5db;'>
                <div style='font-size: 3rem; margin-bottom: 1rem;'></div>
                <h3 style='color: #374151; margin-bottom: 1rem;'>Select Two Players to Compare</h3>
                <p style='margin: 0;'>Choose two players from the dropdowns above to see their detailed comparison.</p>
            </div>
            """,
            label="Player Comparison"
        )

        with gr.Row():
            gr.HTML("""
            <div style="text-align: center; width: 100%; margin: 2rem 0;">
                <h3 style="color: #1E3A8A; margin-bottom: 1rem;"> Performance Radar Charts</h3>
            </div>
            """)

        with gr.Row():
            # Batting Radar Chart
            with gr.Column():
                gr.HTML("""
                <div style="text-align: center; margin-bottom: 1rem;">
                    <h4 style="color: #059669;"> Batting Performance</h4>
                </div>
                """)
                batting_radar_chart = gr.Plot(value=None)

            # Bowling Radar Chart
            with gr.Column():
                gr.HTML("""
                <div style="text-align: center; margin-bottom: 1rem;">
                    <h4 style="color: #DC2626;"> Bowling Performance</h4>
                </div>
                """)
                bowling_radar_chart = gr.Plot(value=None)

        # Event handlers for comparison
        def update_comparison(player1, player2):
            comparison_html, batting_fig, bowling_fig = compare_players(player1, player2)
            return comparison_html, batting_fig, bowling_fig

        # Update comparison when either player selection changes
        player1_dropdown.change(
            fn=update_comparison,
            inputs=[player1_dropdown, player2_dropdown],
            outputs=[comparison_display, batting_radar_chart, bowling_radar_chart]
        )

        player2_dropdown.change(
            fn=update_comparison,
            inputs=[player1_dropdown, player2_dropdown],
            outputs=[comparison_display, batting_radar_chart, bowling_radar_chart]
        )

    return player_comparison

def create_team_recommendation():
    """Create the Team Recommendation tab - Simple category selection with stats"""

    with gr.Column() as team_recommendation:
        # Header
        gr.HTML("""
        <div style="text-align: center; padding: 2rem; background: #f0f9ff; color: white; border-radius: 10px; margin-bottom: 2rem;">
            <h1> Team Recommendation</h1>
            <p>Select a category and get top player recommendations with detailed stats</p>
        </div>
        """)

        # Selection Interface
        with gr.Row():
            with gr.Column():
                category_dropdown = gr.Dropdown(
                    choices=[
                        "PowerPlay Batters",
                        "Middle Overs Batters",
                        "Death Overs Hitters",
                        "Opening Bowlers",
                        "Middle Overs Bowlers",
                        "Death Overs Bowlers",
                        "Wicket Keepers"
                    ],
                    label=" Select Player Category",
                    info="Choose the type of players you want recommendations for"
                )

            with gr.Column():
                num_players_slider = gr.Slider(
                    minimum=1,
                    maximum=15,
                    step=1,
                    value=5,
                    label=" Number of Players",
                    info="How many player recommendations do you want?"
                )

        # Generate Button
        get_players_button = gr.Button(" Get Player Recommendations", variant="primary", size="lg")

        # Results Display
        results_display = gr.HTML(
            value="""
            <div style='text-align: center; padding: 3rem; color: #6B7280; background: #f9fafb; border-radius: 10px; border: 2px dashed #d1d5db;'>
                <div style='font-size: 3rem; margin-bottom: 1rem;'></div>
                <h3 style='color: #374151; margin-bottom: 1rem;'>Select Category & Get Recommendations</h3>
                <p style='margin: 0;'>Choose a player category and number of recommendations above, then click the button.</p>
            </div>
            """
        )

        def get_player_recommendations(category, num_players):
            """Get player recommendations based on category and number"""

            # Map category to dataframe and display config with custom columns
            category_mapping = {
                "PowerPlay Batters": {
                    "df_name": "good_PP_batters_kmeans",
                    "emoji": "",
                    "color": "#059669",
                    "columns": [
                        ("Batter_Runs_PP", "PP Runs", "integer"),
                        ("Batter_Strike_Rate_PP", "PP Strike Rate", "decimal"),
                        ("Batter's_Average", "Average", "decimal"),
                        ("Boundary_Percentage", "Boundary %", "percentage"),
                        ("30+ scores", "30+ Scores", "integer")
                    ]
                },
                "Middle Overs Batters": {
                    "df_name": "good_middle_batters_kmeans",
                    "emoji": "",
                    "color": "#1E3A8A",
                    "columns": [
                        ("Batter_Runs_Middle", "Middle Overs Runs", "integer"),
                        ("Batter's_Average", "Average", "decimal"),
                        ("Batter_Strike_Rate", "Strike Rate", "decimal"),
                        ("Boundary_Percentage", "Boundary %", "percentage"),
                        ("30+ scores", "30+ Scores", "integer")
                    ]
                },
                "Death Overs Hitters": {
                    "df_name": "good_death_overs_hitters_kmeans",
                    "emoji": "",
                    "color": "#DC2626",
                    "columns": [
                        ("Batter_Runs_Death", "Death Overs Runs", "integer"),
                        ("Batter_Strike_Rate_Death", "Death Overs Strike Rate", "decimal"),
                        ("Batter's_Average", "Average", "decimal"),
                        ("Boundary_Percentage", "Boundary %", "percentage"),
                        ("30+ scores", "30+ Scores", "integer")
                    ]
                },
                "Opening Bowlers": {
                    "df_name": "good_T20_opening_bowlers_kmeans",
                    "emoji": "",
                    "color": "#7C3AED",
                    "columns": [
                        ("Wickets_Taken_PP", "PP Wickets", "integer"),
                        ("Bowling_Economy_Rate_PP", "PP Economy", "decimal"),
                        ("Bowling_Average", "Bowling Average", "decimal"),
                        ("Bowling_Strike_Rate", "Bowling Strike Rate", "decimal"),
                        ("3+_Wickets_Haul", "3+ Wicket Hauls", "integer")
                    ]
                },
                "Middle Overs Bowlers": {
                    "df_name": "Middle_T20_Bowlers_cluster_kmeans",
                    "emoji": "",
                    "color": "#EA580C",
                    "columns": [
                        ("Wickets_Taken_Middle", "Middle Overs Wickets", "integer"),
                        ("Bowling_Economy_Rate_Middle", "Middle Overs Economy", "decimal"),
                        ("Bowling_Average", "Bowling Average", "decimal"),
                        ("Bowling_Strike_Rate", "Bowling Strike Rate", "decimal"),
                        ("3+_Wickets_Haul", "3+ Wicket Hauls", "integer")
                    ]
                },
                "Death Overs Bowlers": {
                    "df_name": "death_T20_bowlers_kmeans",
                    "emoji": "",
                    "color": "#B91C1C",
                    "columns": [
                        ("Wickets_Taken_Death", "Death Overs Wickets", "integer"),
                        ("Bowling_Economy_Rate_Death", "Death Overs Economy", "decimal"),
                        ("Bowling_Average", "Bowling Average", "decimal"),
                        ("Bowling_Strike_Rate", "Bowling Strike Rate", "decimal"),
                        ("3+_Wickets_Haul", "3+ Wicket Hauls", "integer")
                    ]
                },
                "Wicket Keepers": {
                    "df_name": "wk_df",
                    "emoji": "",
                    "color": "#059669",
                    "columns": [
                        ("Catches Taken", "Catches", "integer"),
                        ("Run Outs", "Run Outs", "integer"),
                        ("Stumpings", "Stumpings", "integer"),
                        ("Total Dismissals", "Total Dismissals", "integer")
                    ]
                }
            }

            if not category:
                return """
                <div style="text-align: center; padding: 2rem; color: #DC2626;">
                    <h3> Please select a category</h3>
                </div>
                """

            try:
                config = category_mapping[category]
                df_name = config["df_name"]
                emoji = config["emoji"]
                color = config["color"]
                columns = config["columns"]

                # Load the appropriate dataframe
                if df_name == "good_PP_batters_kmeans":
                    df = good_PP_batters_kmeans
                elif df_name == "good_middle_batters_kmeans":
                    df = good_middle_batters_kmeans
                elif df_name == "good_death_overs_hitters_kmeans":
                    df = good_death_overs_hitters_kmeans
                elif df_name == "good_T20_opening_bowlers_kmeans":
                    df = good_T20_opening_bowlers_kmeans
                elif df_name == "Middle_T20_Bowlers_cluster_kmeans":
                    df = Middle_T20_Bowlers_cluster_kmeans
                elif df_name == "death_T20_bowlers_kmeans":
                    df = death_T20_bowlers_kmeans
                elif df_name == "wk_df":
                    df = wk_df

                # Get top N players
                selected_players = df.head(num_players)

                # Create results HTML based on category type
                results_html = f"""
                <div style="background: #f8fafc; padding: 2rem; border-radius: 10px;">
                    <h3 style="color: {color}; text-align: center; margin-bottom: 2rem;">{emoji} {category} - Top {num_players} Recommendations</h3>

                    <div style="background: white; padding: 1.5rem; border-radius: 8px; border-left: 4px solid {color};">
                        <div style="overflow-x: auto;">
                            <table style="width: 100%; border-collapse: collapse; font-size: 0.9rem;">
                                <thead>
                                    <tr style="background: #f3f4f6;">
                                        <th style="padding: 0.75rem; text-align: left; border: 1px solid #e5e7eb;">Rank</th>
                                        <th style="padding: 0.75rem; text-align: left; border: 1px solid #e5e7eb;">Player Name</th>
                """

                # Add custom headers for this category
                for _, header, _ in columns:
                    results_html += f'<th style="padding: 0.75rem; text-align: center; border: 1px solid #e5e7eb;">{header}</th>'

                results_html += """
                                    </tr>
                                </thead>
                                <tbody>
                """

                # Add player rows with custom stats
                for idx, (player_name, player_data) in enumerate(selected_players.iterrows(), 1):
                    # Handle different ways player names might be stored
                    if 'Player' in selected_players.columns:
                        display_name = player_data['Player']
                    else:
                        display_name = player_name

                    results_html += f"""
                                    <tr>
                                        <td style="padding: 0.75rem; border: 1px solid #e5e7eb; font-weight: bold;">{idx}</td>
                                        <td style="padding: 0.75rem; border: 1px solid #e5e7eb;"><strong>{display_name}</strong></td>
                    """

                    # Add custom stats for this category
                    for col_name, _, format_type in columns:
                        value = player_data.get(col_name, 0)

                        if format_type == "integer":
                            formatted_value = f"{value:,.0f}"
                        elif format_type == "decimal":
                            formatted_value = f"{value:.2f}"
                        elif format_type == "percentage":
                            formatted_value = f"{value:.1f}%"
                        else:
                            formatted_value = f"{value}"

                        results_html += f'<td style="padding: 0.75rem; text-align: center; border: 1px solid #e5e7eb;">{formatted_value}</td>'

                    results_html += "</tr>"

                results_html += """
                                </tbody>
                            </table>
                        </div>
                    </div>
                </div>
                """

                return results_html

            except Exception as e:
                return f"""
                <div style="text-align: center; padding: 2rem; color: #DC2626;">
                    <h3> Error loading recommendations</h3>
                    <p>Error: {str(e)}</p>
                    <p style="font-size: 0.9rem; color: #6B7280;">Please ensure the dataframes are loaded correctly.</p>
                </div>
                """

        # Connect button to function
        get_players_button.click(
            fn=get_player_recommendations,
            inputs=[category_dropdown, num_players_slider],
            outputs=[results_display]
        )

    return team_recommendation

def create_main_interface():
    """Create the main interface with tabs"""

    with gr.Blocks(
        theme=gr.themes.Soft(),
        title="Cricklytics - Cricket Analytics Platform",
        css="""
        /* Remove max-width constraint to use full screen */
        .gradio-container {
            width: 100% !important;
            max-width: none !important;
            font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
        }

        /* Make the main container full width */
        .main {
            width: 100% !important;
            max-width: none !important;
        }

        /* Ensure content spans full width */
        .container {
            width: 100% !important;
            max-width: none !important;
            padding: 0 20px;
        }

        /* Custom tab styling to match the blue theme */
        .tab-nav button {
            color: #374151 !important;
        }

        .tab-nav button.selected {
            color: #1E3A8A !important;
            border-bottom-color: #1E3A8A !important;
        }

        /* Ensure consistent styling */
        .gradio-html {
            margin: 0 !important;
        }

        /* Full width for all components */
        .block {
            width: 100% !important;
        }
        """
    ) as main_interface:

        with gr.Tabs() as tabs:

            # Home Page Tab
            with gr.Tab(" Home", id="home"):
                create_home_page()

            # Player Overview Tab
            with gr.Tab(" Player Overview", id="player_overview"):
                create_player_overview()

            # Player Comparison Tab
            with gr.Tab(" Player Comparison", id="player_comparison"):
                create_player_comparison()

            # Team Recommendation Tab
            with gr.Tab(" Team Recommendation", id="team_recommendation"):
                create_team_recommendation()

    return main_interface

# Launch the application
if __name__ == "__main__":
    app = create_main_interface()
    app.launch()
```
